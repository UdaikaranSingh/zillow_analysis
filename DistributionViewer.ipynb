# Distribution Viewer Notebook

### Purpose of this notebook is to see the images of the distributions



Possible Distributions:
1. primary_date_dist
2. conditional_dist


```python
import os
from IPython.display import Image
from constants import FIPS, FIP_to_state
```

```python
gen_directory = "/Volumes/GoogleDrive/Shared drives/Zillow1/data/gen/bystate"
#note: if using the Drive data, set this to the by_state directory
```

### This method is getting the Distributions for a certain folder for a specific state

```python
distributions = "primary_date_dist"
state = "02"
folder_with_images = os.path.join(os.path.join(gen_directory,state), distributions)
for img_name in os.listdir(folder_with_images):
    pth = os.path.join(folder_with_images, img_name)
    display(Image(filename=pth))
    print("----------------------------------------------------------------------------")
```

### This method is getting a specific Distribution for over all States

```python
distributions = "primary_date_dist"
file_name = "primary_dates_hist_5.png"

for state in FIPS:
    folder_with_images = os.path.join(os.path.join(gen_directory,state), distributions)
    file_path = os.path.join(folder_with_images, file_name)
    if os.path.exists(file_path):
        print("State: {}".format(FIP_to_state[state]))
        display(Image(filename=file_path))
        print("----------------------------------------------------------------------------")
```
